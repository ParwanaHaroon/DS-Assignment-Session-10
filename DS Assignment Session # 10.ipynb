```python

import numpy as np
import pandas as pd
from scipy import optimize
import matplotlib.pyplot as plt
get_ipython().run_line_magic('matplotlib', 'inline')

```

```python
months = list(range(12))
maxTemps = [39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25]
minTemps = [21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18]

def yearly_temps(times, avg, ampl, time_offset):
    return (avg + ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temps, months, maxTemps, [20, 10, 0])
res_min, cov_min = optimize.curve_fit(yearly_temps, months, minTemps, [-40, 20, 0])


days = np.linspace(0, 12, num = 365)
plt.figure()
plt.plot(months, maxTemps, 'ro')
plt.plot(days, yearly_temps(days, *res_max), 'r-' )
plt.plot(months, minTemps, 'bo')
plt.plot(days, yearly_temps(days, * res_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')
plt.show()
```

```python
# Create a scatterplot with the fare paid and age, differ the plot color by gender
```

```python
def selectColor(x):
    if(x == 'female'):
        return 'Blue'
    else:
        return 'Red'

titanicDS['colors'] = 'Red'
titanicDS['colors'] = titanicDS['sex'].apply(selectColor)

plt.scatter(titanicDS['age'], titanicDS['fare'], c=titanicDS['colors'])
```
